**Goal:** Build a small `AnnData` from Xenium *preview* bundle, then run a minimal Scanpy pipeline.
We will **avoid the HDF5 matrix** and build counts from `transcripts.parquet` to keep RAM bounded.

```python
# schemas; these help us map column names/types on your bundle
cells_pq = FOLDER / "cells.parquet"
tx_pq    = FOLDER / "transcripts.parquet"

cells_schema = pq.read_schema(cells_pq)
tx_schema    = pq.read_schema(tx_pq)

print(cells_schema, tx_schema)
```

```python
# figure out the gene column name in transcripts.parquet
tx_cols = set(tx_schema.names)
if "gene" in tx_cols:
    GENE_COL = "gene"
elif "feature_name" in tx_cols:
    GENE_COL = "feature_name"
else:
    raise KeyError(f"Cannot find gene column in transcripts.parquet; saw columns: {tx_cols}")

GENE_COL

```

```python
MAX_CELLS = 10000000  # tune down if RAM is tight (e.g., 10000)

cells_tbl = pq.read_table(cells_pq)
cells_df  = cells_tbl.to_pandas()

# Enforce string IDs
if "cell_id" not in cells_df.columns:
    raise KeyError("cells.parquet missing `cell_id` column")
cells_df["cell_id"] = to_str_series(cells_df["cell_id"])

n_all = len(cells_df)
cells_df = cells_df.sample(n=min(MAX_CELLS, n_all), random_state=42).reset_index(drop=True)
keep_cell_ids = set(cells_df["cell_id"].tolist())
n_all, len(cells_df)
```

```python
# figure out the gene column name in transcripts.parquet
tx_cols = set(tx_schema.names)
if "gene" in tx_cols:
    GENE_COL = "gene"
elif "feature_name" in tx_cols:
    GENE_COL = "feature_name"
else:
    raise KeyError(f"Cannot find gene column in transcripts.parquet; saw columns: {tx_cols}")

GENE_COL
```

```python
# We scan transcripts.parquet in batches, restricted to our sampled cell_ids

dataset = ds.dataset(tx_pq)
# If the dataset stored cell_id as binary, we need a bytes set to filter efficiently
# Build a mapping: string_id -> bytes_id from the transcripts schema type
# Simpler fallback: filter everything and drop non-sampled rows in pandas (slower but robust).

# Try fast path: bytes filter
use_fast_filter = False
try:
    # Probe a tiny batch to see dtype
    tiny = next(iter(dataset.to_batches(columns=["cell_id"], batch_size=1000)))
    cell_id_is_binary = tiny.column(0).type == pa.binary()
    use_fast_filter = cell_id_is_binary
except StopIteration:
    cell_id_is_binary = False

if use_fast_filter:
    keep_cell_ids_bytes = [s.encode("utf-8") for s in keep_cell_ids]
    filt = ds.field("cell_id").isin(keep_cell_ids_bytes)
    batches = dataset.to_batches(columns=["cell_id", GENE_COL], filter=filt, batch_size=2_000_000)
else:
    # robust fallback (reads more, but still batched)
    batches = dataset.to_batches(columns=["cell_id", GENE_COL], batch_size=1_000_000)

# Count per (cell_id, gene)
counts = {}
gene_totals = {}

for bi, batch in enumerate(batches, start=1):
    df = batch.to_pandas()
    # normalize types to str
    df["cell_id"] = to_str_series(df["cell_id"])
    df[GENE_COL]  = to_str_series(df[GENE_COL])
    if not use_fast_filter:
        df = df[df["cell_id"].isin(keep_cell_ids)]
    # group and add to tallies
    g = df.groupby(["cell_id", GENE_COL]).size().astype(np.int32)
    for (cid, gene), n in g.items():
        d = counts.setdefault(cid, {})
        d[gene] = d.get(gene, 0) + int(n)
        gene_totals[gene] = gene_totals.get(gene, 0) + int(n)
    del df, g; gc.collect()
    if bi % 5 == 0:
        print(f"[batches] processed: {bi}")

len(counts), len(gene_totals)
```

```python
MAX_GENES = 2000  # tune; 1000 is fine for a quick run

top_genes = sorted(gene_totals, key=gene_totals.get, reverse=True)[:MAX_GENES]
gi = {g:i for i,g in enumerate(top_genes)}

# Keep ordered cell_ids (intersection with sampled cells)
ordered_ids = sorted(counts.keys())
ci = {c:i for i,c in enumerate(ordered_ids)}

rows, cols, data = [], [], []
for cid, gdict in counts.items():
    r = ci[cid]
    for g, n in gdict.items():
        j = gi.get(g)
        if j is None: 
            continue
        rows.append(r); cols.append(j); data.append(n)

X = coo_matrix((data, (rows, cols)), shape=(len(ordered_ids), len(top_genes)), dtype=np.int32).tocsr()
X.shape, X.nnz
```

```python
# Reindex cells_df to match matrix order
obs = cells_df.set_index("cell_id").reindex(ordered_ids)
# Create AnnData
adata = ad.AnnData(X=X, obs=obs, var=pd.DataFrame(index=pd.Index(top_genes, name="gene")))
adata
```

```python
# Light normalization & clustering
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, n_top_genes=min(1500, adata.n_vars), subset=True)
sc.pp.pca(adata, n_comps=50)
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)
sc.tl.umap(adata)
sc.tl.leiden(adata, resolution=0.4)

adata
```

```python
sc.pl.umap(adata, color="leiden", legend_loc="on data", size=8)
plt.savefig(FIGS / "umap_leiden.png", dpi=180); plt.close()

for g in ["EPCAM", "PTPRC", "PECAM1"]:
    if g in adata.var_names:
        sc.pl.umap(adata, color=g, cmap="viridis", size=8)
        plt.savefig(FIGS / f"umap_marker_{g}.png", dpi=180); plt.close()

summary = f"""# Quick Summary — {datetime.now().isoformat(timespec='seconds')}
- Source: {FOLDER.name}
- Cells: {adata.n_obs}
- Genes: {adata.n_vars}
- Leiden clusters: {adata.obs['leiden'].nunique()}
"""
(RESULTS / "summary_preview.md").write_text(summary)
summary
```
